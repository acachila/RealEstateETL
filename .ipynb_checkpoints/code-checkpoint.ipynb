```python
import pandas as pd
from datetime import datetime
import datetime as dt
```

```python
data = pd.read_csv("Zip_MedianListingPricePerSqft_AllHomes.csv")
data.head()
```

```python
# start = input("start: ")
# end = input("end: ")
start = "2015-2-1"
end = "2015-6-1"
dates = pd.DataFrame({"md": pd.date_range(start=start, end=end)})
# dates["md"] = dates["md"].dt.strftime("%Y-%m-%d")
months = dates["md"].dt.strftime("%Y-%m")
months = months.unique()
months
```

```python
# start = input("start: ")
# end = input("end: ")
start = "2015-2-1"
end = "2015-6-1"
start = datetime.strptime(start,'%Y-%m-%d')
end = datetime.strptime(end,'%Y-%m-%d')
start = datetime.strftime(start,'%Y-%m-%d')
end = datetime.strftime(end,'%Y-%m-%d')
df = data[["RegionName","City","State","Metro","CountyName",start,end]]
df["diff"] = df[end]/df[start]
df.sort_values("diff",ascending=False)
```
